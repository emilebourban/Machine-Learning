```python
def to_float(x): 
    for k in range (2,len(x[0])):
        for j in range(1,len(x)):
            x[j][k] = float(x[j][k])
    return None
to_float(data)
```

```python
#counts
def count(x,k,liste1):
    h = 0
    for j in range(1, len(x)):
        if x[j][k] == -999.0 :
            h += 1
    liste1.append(h)

def column_count(x):
    liste1 = []
    for k in range(2,len(x[0])):
        count(x,k,liste1)
    print(len(liste1))
    for k in range (2,len(x[0])):
        print(x[0][k], liste1[k-2]*100/len(x))

column_count(data)
```
